## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
try:
    print("Missing Values Before:\n")
    print(df.isnull().sum())
    df.drop_duplicates(inplace=True)

    # Standardize column names
    df.columns = df.columns.str.lower().str.replace(' ', '_')

    # Handle missing values
    df['shipment_mode'].fillna(df['shipment_mode'].mode()[0], inplace=True)
    df['line_item_insurance_(usd)'].fillna(df['line_item_insurance_(usd)'].median(), inplace=True)
    df['dosage'].fillna('Unknown', inplace=True)

    print("\nMissing Values After:\n")
    print(df.isnull().sum())
    print(f"\nDataset shape: {df.shape}")
    print("\nData Wrangling Completed Successfully")

except Exception as e:
    print(f"Error during Data Wrangling: {e}")
````

### What all manipulations have you done and insights you found?

Missing values were identified. Dupicate records were checked and removed to improve data consistency. The columns were standarised to improve readablity.
Insights- there were only 3 columns with missing values. ````python

````python
df['scheduled_delivery_date'] = pd.to_datetime(df['scheduled_delivery_date'])
df['delivered_to_client_date'] = pd.to_datetime(df['delivered_to_client_date'])
df['Delivery Status'] = df.apply(
    lambda x: 'On-Time'
    if x['delivered_to_client_date'] <= x['scheduled_delivery_date']
    else 'Late',
    axis=1
)
grouped = df.groupby(['country', 'Delivery Status']).size().unstack(fill_value=0)
grouped = grouped.sort_values(by='Late',ascending=False).head(10)
grouped.plot(kind='bar',stacked=True,figsize=(12,6))
plt.title('On-Time vs Late Delivery by Country')
plt.xlabel('Country')
plt.ylabel('Number of Deliveries')
plt.xticks(rotation=45)
plt.legend(title='Delivery Status')
plt.show()
````
